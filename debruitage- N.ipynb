```python
import numpy as np
import scipy.spatial
import matplotlib.pyplot as plt
from skimage import io, img_as_ubyte, img_as_float
from skimage.restoration import denoise_nl_means, estimate_sigma, denoise_bilateral
```

```python
fichier = np.fromfile("F2.raw", np.uint8)
```

```python
fichier = fichier.reshape((1080, 600, 1050))
```

```python
def mean(img):
    m=img.ravel()
    k=10
    r=[]
    for i in range(len(m)-k):
        r.append(int(sum(m[i:i+k])/k))
    r=np.concatenate([r,m[-k:]])
    return r.reshape((600,1050))
```

```python
h=mean(fichier[100])
```

```python
plt.imshow(h,cmap='gray')
```
