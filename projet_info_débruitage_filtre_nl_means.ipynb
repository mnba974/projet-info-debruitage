```python
import numpy as np
import scipy.spatial
import matplotlib.pyplot as plt
from skimage import io, img_as_ubyte, img_as_float
from skimage.restoration import denoise_nl_means, estimate_sigma, denoise_bilateral
from tqdm import tqdm
```

```python
fichier = np.fromfile("photo1.raw", np.uint8)
```

```python
print(fichier.shape)
```

```python
fichier = fichier.reshape((1080, 600, 1050))
```

```python
for i in range(0, len(fichier), 108) : 
    plt.imshow(fichier[i])
    plt.show()
```

```python
image_test = fichier[108]
image_test.shape
```

```python
plt.imshow(image_test, cmap="gray")
```

```python
# Estimer le sigma du bruit
sigma_est = np.mean(estimate_sigma(image_test))

# Appliquer Non-Local Means Denoising
denoised_image = denoise_nl_means(image_test, h=0.1 * sigma_est, fast_mode=True)

plt.imshow(denoised_image, cmap='gray')
plt.show()
plt.imshow(image_test, cmap='gray')
plt.show()


```

```python
#nlm qu'on utilise
for i in range(0, len(fichier), 59) : 
    
    image_array = img_as_ubyte(fichier[i])

    # Estimer le sigma du bruit
    sigma_est = np.mean(estimate_sigma(image_array))

    # Appliquer Non-Local Means Denoising
    denoised_im = denoise_nl_means(image_array, h=0.1 * sigma_est, fast_mode=True)

    plt.imshow(denoised_im, cmap='gray')
    plt.show()
```
